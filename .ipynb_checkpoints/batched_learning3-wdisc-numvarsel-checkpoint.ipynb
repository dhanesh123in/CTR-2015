```python
from sklearn.naive_bayes import BernoulliNB as bnb
from sklearn.linear_model import PassiveAggressiveClassifier as pac
from sklearn.linear_model import SGDClassifier as sgd
from sklearn.linear_model import Perceptron as pcp
import pandas as pd
import numpy as np
import datetime as dt
from sklearn.metrics import log_loss
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score as auc
from sklearn.svm import SVC

from sklearn.feature_extraction import DictVectorizer
from sklearn.ensemble import RandomForestClassifier as rfc

#Utility function to convert numeric variable to quantile level string
def qfun(x,y,cname):
    for i in range(len(y)):
        if x<=y[y.index[i]]:
            return(cname+'_'+str(i))
    return(cname+'_'+str(i))
```

```python
#Discretizaton of numeric variables to quantiles
def numvardisc(Xd,numq=5,rtype='train',qmodel=None):
    msk=Xd.dtypes<>'object'
    
    if rtype=='train':
        qlevdf=Xd.loc[:,msk].quantile(np.array(range(1,numq+1)).astype('float')*(1.0/float(numq)))
    else:
        qlevdf=qmodel
        
    for i in range(len(Xd.loc[:,msk].columns)):
        colname=Xd.loc[:,msk].columns.values[i]
        #Xd[colname]=Xd[colname].apply(qfun,1,y=qlevdf[colname],cname='q')
        Cd=pd.Series(Xd[colname])
        y=qlevdf[colname]
        
        for j in range(len(y)):
            if j==0:
                Cd[Xd[colname]<=y[y.index[j]]]='q_'+str(j)
            elif j<>(len(y)-1):
                Cd[(Xd[colname]>y[y.index[j-1]])&(Xd[colname]<=y[y.index[j]])]='q_'+str(j)
            else:
                Cd[(Xd[colname]>y[y.index[j-1]])]='q_'+str(j)
        
        Xd[colname]=Cd
        
    return dict(Xd=Xd,qlevdf=qlevdf)
            
#This is unsupervised really at present   
def discvarsel(Xd,Y,supp=20,rtype='train',dmodel=None):
    msk=Xd.dtypes=='object'
    oplev=list()

    for i in range(len(Xd.loc[:,msk].columns)):
        colname=Xd.loc[:,msk].columns.values[i]
        #print colname
        
        if rtype=='train':
            xx=pd.DataFrame(pd.crosstab(Xd[colname],Y))
            #print xx.columns.values
            #Compute conversion and number of points in each level
            xx['conv']=xx[1]/(xx[1]+xx[0])
            xx['total']=xx[0]+xx[1]
            
            if i==0:
                #Compute base conversion and weight of evidence
                bconv=np.sum(xx[1]).astype('float64')/(np.sum(xx[0]).astype('float64')+np.sum(xx[1]).astype('float64'))
                bwoe=np.log(bconv/(1-bconv))
                #print bconv

            xx['woe']=np.log(xx['conv']/(1-xx['conv']))-bwoe

            #Find infrequent levels
            xy=xx.loc[xx['total']<supp,0].index
            oplev.append(xy)
            #print xy
        else:
            oplev=dmodel
            xy=oplev[i]
        
        #Rename infrequent levels to 'Others'
        rind=Xd[colname].isin(xy)
        Xd[colname].loc[rind]='Others'
        #print Xd.shape
        
    #print oplev
    return dict(Xd=Xd,oplev=oplev)
```

```python
colnames=['C1','banner_pos','site_id','site_domain','site_category','app_id','app_domain','app_category','device_id','device_ip','device_model','device_type','device_conn_type','C14','C15','C16','C17','C18','C19','C20','C21','day','numv1','numv2','numv3','numv4','numv5','numv6','numv7','numv8','numv9','numv10','numv11','numv12','weekday','hr']
numvars=['numv1','numv2','numv3','numv4','numv5','numv6','numv7','numv8','numv9','numv10','numv11','numv12']

vec = DictVectorizer(sparse=True)

#Batched reading and learning
batch=0
nlines=0
nlim=200000
blim=5
btrain=5
X=np.zeros((nlim,39),dtype='a20')
#supp=0.001*nlim #0.1%
supp=50
ll=0.0
bconv=0.0
numq=3

#clfs=[dtc(random_state=0,min_samples_leaf=50) for i in range(btrain)]
clf = sgd(random_state=123,loss='log')
vec = DictVectorizer(sparse=True)

with open("train_p.csv","r") as f:
    for line in f:
        
        X[nlines,:37]=[str(xx.strip()) for xx in line.strip().split(',')]
        #print X[nlines,:]
        X[nlines,37]=str(dt.datetime.strptime(X[nlines,2],'%y%m%d%H').weekday())
        X[nlines,38]=X[nlines,2][-2:]
        nlines=nlines+1;
        if (nlines==nlim):
            print 'Processing batch '+str(batch+1)
            batch=batch+1

            Xd=pd.DataFrame(X[:,3:],columns=colnames)
            #Xd.drop('device_id', axis=1, inplace=True)
            #Xd.drop('device_ip', axis=1, inplace=True)
            #Xd.drop('device_model', axis=1, inplace=True)
            #Xd.drop('site_id', axis=1, inplace=True)
            #Xd.drop('site_domain', axis=1, inplace=True)
            #Xd.drop('app_id', axis=1, inplace=True)
            #Xd.drop('app_domain', axis=1, inplace=True)
            for nv in numvars:
                Xd[nv]=Xd[nv].astype('int')
            
            Y=X[:,1].astype('int')
            print float(np.sum(Y))/float(nlim)
            #All the jazz goes here
            if (batch==1):
                print 'Running Dict Vectorizer fit transform'
               
                XX=numvardisc(Xd,numq=numq,rtype='train')
                Xd=XX['Xd']
                qlevdf=XX['qlevdf']
                del XX
                #print XX
                
                XY=discvarsel(Xd,Y,supp=supp,rtype='train')
                Xd=XY['Xd']
                oplev=XY['oplev']
                del XY
                #print XY
                
                X_sparse=vec.fit_transform(Xd.T.to_dict().values())
                clf.fit(X_sparse,Y)
            else:
                print 'Running Dict Vectorizer transform alone'
                XX=numvardisc(Xd,numq=numq,rtype='predict',qmodel=qlevdf) 
                Xd=XX['Xd']
                del XX                
                
                XY=discvarsel(Xd,Y,supp=supp,rtype='predict',dmodel=oplev)
                Xd=XY['Xd']
                del XY
                
                X_sparse=vec.transform(Xd.T.to_dict().values())

            
            ypred=clf.predict_proba(X_sparse)[:,1]
            if (batch>1):
                ll=ll+log_loss(Y,ypred)
                bconv=bconv+float(np.sum(Y))/float(nlim)
                bconva=bconv/float(batch-1)
                print ll/float(batch-1),-(bconva*np.log(bconva)+(1-bconva)*np.log(1-bconva))
            else:
                bconva=float(np.sum(Y))/float(nlim)
                print log_loss(Y,ypred),-(bconva*np.log(bconva)+(1-bconva)*np.log(1-bconva))
            
            nlines=0
            X=np.zeros((nlim,39),dtype='a20')
            #del X_sparse
            
            if (batch>=blim):
                print "Done.. aren't we?"
                f.close()
                del X
                break

print 'Yes, we are..'
```

```python
##Prediction

batch=0
nlines=0
nlim=100000
#g=open("first_submission.csv",'a')
g=open("tenth_submission.csv",'a')
g.write("id,click\n")

Yp=np.zeros((nlim,2),dtype='a20')
X=np.zeros((nlim,38),dtype='a20')
with open("test_p.csv","r") as f:
    for line in f:
        
        X[nlines,:36]=[str(xx.strip()) for xx in line.strip().split(',')]
        #print X[nlines,:]
        X[nlines,36]=str(dt.datetime.strptime(X[nlines,1],'%y%m%d%H').weekday())
        X[nlines,37]=X[nlines,1][-2:]
        nlines=nlines+1;
        if (nlines==nlim):
            print 'Processing batch '+str(batch+1)
            batch=batch+1

            Xd=pd.DataFrame(X[:,2:],columns=colnames)
            
            for nv in numvars:
                Xd[nv]=Xd[nv].astype('int')            

            XX=numvardisc(Xd,numq=numq,rtype='predict',qmodel=qlevdf) 
            Xd=XX['Xd']
            del XX                

            XY=discvarsel(Xd,Y,supp=supp,rtype='predict',dmodel=oplev)
            Xd=XY['Xd']
            del XY                
                
            X_sparse=vec.transform(Xd.T.to_dict().values())                        
            ypred=clf.predict_proba(X_sparse)[:,1]
            
            #Output to file
            Yp[:,0]=X[:,0]
            Yp[:,1]=[str(xx) for xx in ypred]
            np.savetxt(g,Yp,delimiter=',',fmt="%s")
            
            nlines=0
            X=np.zeros((nlim,38),dtype='a20')
            del X_sparse
            
    else:
        print "Processing final batch.."
        
        Xd=pd.DataFrame(X[:nlines,2:],columns=colnames)
        
        for nv in numvars:
            Xd[nv]=Xd[nv].astype('int')

        XX=numvardisc(Xd,numq=numq,rtype='predict',qmodel=qlevdf) 
        Xd=XX['Xd']
        del XX                

        XY=discvarsel(Xd,Y,supp=supp,rtype='predict',dmodel=oplev)
        Xd=XY['Xd']
        del XY            
            
        X_sparse=vec.transform(Xd.T.to_dict().values())                        
        ypred=clf.predict_proba(X_sparse)[:,1]

        #Output to file
        Yp=np.zeros((nlines,2),dtype='a20')
        Yp[:,0]=X[:nlines,0]
        Yp[:,1]=[str(xx) for xx in ypred]
        np.savetxt(g,Yp,delimiter=',',fmt="%s")
        
        f.close()
        del X
        del X_sparse

print 'Yes, we are done..'
g.close()
```
